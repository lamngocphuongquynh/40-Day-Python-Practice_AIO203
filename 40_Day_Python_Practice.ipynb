```python
def calculate_can_chi_calendar(year):
    list_can = ["Canh", "Tân","Nhâm","Quý","Giáp","Ất","Bính","Đinh","Mậu","Kỷ"]
    list_chi = ["Thân","Dậu","Tuất","Hợi","Tý","Sửu","Dần","Mẹo","Thìn","Tỵ","Ngọ","Mùi"]
    for i in range(len(list_can)):
        for j in range(len(list_chi)):
            if year%10 == i and year%12 == j:
                return list_can[i] + " " + list_chi[j]
```

```python
import math

def quadratic_equation(a,b,c):
    delta = b*b - 4*a*c
    if delta > 0:
        x1 = (-b - math.sqrt(delta))/(2*a)
        x2 = (-b + math.sqrt(delta))/(2*a)
        return x1, x2
    elif delta == 0:
        x1 = -b/(2*a)
        x2 = x1
        return x1, x2
    else:
        return "phương trình vô nghiệm"
```

```python
corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"]
dictionary = []

for string in corpus:
    lower_split_string = string.lower().split()
    for i in lower_split_string:
        if i not in dictionary:
            dictionary.append(i)
print(dictionary)
```

```python
my_text = "Tôi thích AI thích Toán"
for text in my_text:
    lower_split_text = my_text.lower().split()
print(lower_split_text)
```

```python
def vectorize(lower_split_text, dictionary):
    word_count_dict = {word: 0 for word in dictionary}
    for token in lower_split_text:
        if token in dictionary:
            word_count_dict[token] += 1
        else:
            pass
    vector = list(word_count_dict.values())
    return vector

vectorize(lower_split_text, dictionary)
```

```python
B = [
    [2, 4, 6],
    [1, 3, 5],
    [1, 0, 1]
]
```

```python
def sum_matrix(X, Y):
    row = len(X)
    column = len(X[0])
    sum_result = [[0]*column for row in range(row)]
    for i in range(row):
        for j in range(column):
             sum_result[i][j] = X[i][j] + Y[i][j]
    return sum_result
```

```python
sum_matrix(A, B)
```

```python
def subtract_matrix(X, Y):
    row = len(X)
    column = len(X[0])
    subtract_result = [[0]*column for row in range(row)]
    for i in range(row):
        for j in range(column):
             subtract_result[i][j] = X[i][j] - Y[i][j]
    return subtract_result
```

```python
subtract_matrix(A, B)
```

```python
def dot_product_matrix(X, Y):
    row = len(X)
    column = len(X[0])
    dot_product_result = [[0]*column for row in range(row)]
    for i in range(row):
        for j in range(len(Y[0])):
            dot_product = 0
            for k in range(len(Y)):
                dot_product += X[i][k] * Y[k][j]
                dot_product_result[i][j] = dot_product
    return dot_product_result
```

```python
dot_product_matrix(A, B)
```

```python
import random

def random_number_with_condition(total):
    random.seed(0)
    count_total = 0
    while True:
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        if a + b == total:
            count_total += 1
            break
        count_total += 1
    return count_total

print(random_number_with_condition(40))
print(random_number_with_condition(20))
print(random_number_with_condition(35))
```

```python
def find_devisible_number(a):
    x = 101
    while True:
        if x%a == 0:
            break
        else:
            x += 1
    return x

print(find_devisible_number(5))
print(find_devisible_number(17))
```

```python
def find_squared_root(a):
    epsilon = 0.001
    x = 1

    while True:
        fx = x*x - a
        dx = 2*x
        x_n = x - (fx/dx)
        if abs(x_n - x) < epsilon:
            break
        x = x_n
    return x_n

print(find_squared_root(2))
print(find_squared_root(3))
```

```python
import numpy as np
```

```python
def gaussian_kernel(size, sigma):
    if size%2 == 0:
        size = size + 1

    max_point = size//2
    min_point = -max_point

    K = np.zeros((size, size))
    for x in range(min_point, max_point + 1):
        for y in range(min_point, max_point + 1):
            value = 1/(2*np.pi*sigma*sigma)*np.exp(-(x*x + y*y)/(2*sigma*sigma))
            K[x - min_point, y - min_point] = value

    return K


kernel_1 = gaussian_kernel(5, 2.5)
img_gaussian_1 = cv2. filter2D(img, -1, kernel_1)
plt.imshow(img_gaussian_1)
plt.title('Sigma=2.5')
plt.show()

kernel_2 = gaussian_kernel(5, 5.0)
img_gaussian_2 = cv2. filter2D(img, -1, kernel_2)
plt.imshow(img_gaussian_2)
plt.title('Sigma=5.0')
plt.show()

kernel_3 = gaussian_kernel(5, 10.0)
img_gaussian_3 = cv2. filter2D(img, -1, kernel_3)
plt.imshow(img_gaussian_3)
plt.title('Sigma=10.0')
plt.show()
```

```python
import pandas as pd

path = '/content/drive/MyDrive/data/advertising.csv'

df = pd.read_csv(path)
df.head()
```

```python
def scaling(data1, data2, data3):
    data = data1 + data2 + data3
    max_value = max(data)
    min_value = min(data)
    mean_value = sum(data) / len(data)

    data1 = [(x - mean_value) / (max_value - min_value) for x in data1]
    data2 = [(x - mean_value) / (max_value - min_value) for x in data2]
    data3 = [(x - mean_value) / (max_value - min_value) for x in data3]
    return (data1, data2, data3)

def predict(x1, x2, x3, w1, w2, w3, b):
    return x1*w1 + x2*w2 + x3*w3 + b

def gradient_w(y_hat, y, xi):
    dwi = 2*xi*(y_hat - y)
    return dwi

def gradient_b(y_hat, y):
    db = 2*(y_hat - y)
    return db

def update_weight_w(wi, dwi, lr):
    wi_new = wi - lr*dwi
    return wi_new

def update_weight_b(b, db, lr):
    b_new = b - lr*db
    return b_new
```

```python
b = 1
w1, w2, w3 = 0, 0, 0
lr = 0.01
epochs = 1000

wandb.init(
    project="demo-linear-regression",
    config={
        "learning_rate": lr,
        "epochs": epochs,
    },
)

wandb.run.log({"Dataset": wandb.Table(dataframe=df)})

tv_data = df['TV']
radio_data = df['Radio']
newspaper_data = df['Newspaper']
(X1_train, X2_train, X3_train) = scaling(tv_data, radio_data, newspaper_data)
Y_train = df['Sales']

N = len(X1_train)
losses = []

for epoch in range(epochs):
    loss_total = 0.0
    for i in range(N):
        x1 = X1_train[i]
        x2 = X2_train[i]
        x3 = X3_train[i]
        y = Y_train[i]

        y_hat = predict(x1, x2, x3, w1, w2, w3, b)
        dw1 = gradient_w(y_hat, y, x1)
        dw2 = gradient_w(y_hat, y, x2)
        dw3 = gradient_w(y_hat, y, x3)
        db = gradient_b(y_hat, y)

        w1 = update_weight_w(w1, dw1, lr)
        w2 = update_weight_w(w2, dw2, lr)
        w3 = update_weight_w(w3, dw3, lr)
        b = update_weight_b(b, db, lr)

        loss = (y_hat - y)*(y_hat - y)/2.0
        loss_total = loss_total + loss
    wandb.log({"loss": loss_total/N})
    losses.append(loss_total/N)

wandb.finish()
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
lst_1 = [[1, -2, 1],
         [-3, 1, 4],
         [-2, 5, 1]]

lst_2 = [[1, 3, 5],
         [2, 4, 6],
         [3, 5, 7]]
```

```python
arr_1 = np.array(lst_1)
arr_2 = np.array(lst_2)

arr_mul = np.multiply(arr_1, arr_2)
arr_div = np.divide(arr_1, arr_2)

print('Mảng 1:\n', arr_1)
print('\nMảng 2:\n', arr_2)
print('\nPhép nhân 2 mảng:\n', arr_mul)
print('\nPhép chia 2 mảng:\n', arr_div)
```

```python
tensor_pt_1 = torch.tensor(lst_1)
tensor_pt_2 = torch.tensor(lst_2)

tensor_pt_mul = torch.multiply(tensor_pt_1, tensor_pt_2)
tensor_pt_div = torch.divide(tensor_pt_1, tensor_pt_2)

print('Tensor 1:\n', tensor_pt_1)
print('\nTensor 2:\n', tensor_pt_2)
print('\nPhép nhân 2 tensor:\n', tensor_pt_mul)
print('\nPhép chia 2 tensor:\n', tensor_pt_div)
```

```python
tensor_tf_1 = tf.constant(lst_1)
tensor_tf_2 = tf.constant(lst_2)

tensor_tf_mul = tf.multiply(tensor_tf_1, tensor_tf_2)
tensor_tf_div = tf.divide(tensor_tf_1, tensor_tf_2)

print('Tensor 1:\n', tensor_tf_1)
print('\nTensor 2:\n', tensor_tf_2)
print('\nPhép nhân 2 tensor:\n', tensor_tf_mul)
print('\nPhép chia 2 tensor:\n', tensor_tf_div)
```

**Các Phép Tính Numpy, Pytorch và Tensorflow - Phần 3**

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
a = [1, 4, 7]
b = [9, 2, 3]
```

```python
arr_1 = np.array(a)
arr_2 = np.array(b)

arr_dot = np.dot(arr_1, arr_2)

print('Mảng 1:\n', arr_1)
print('\nMảng 2:\n', arr_2)
print('\nTích vô hướng 2 mảng:\n', arr_dot)
```

```python
tensor_pt_1 = torch.tensor(a)
tensor_pt_2 = torch.tensor(b)

tensor_pt_dot = torch.dot(tensor_pt_1, tensor_pt_2)

print('Tensor 1:\n', tensor_pt_1)
print('\nTensor 2:\n', tensor_pt_2)
print('\nTích vô hướng 2 tensor:\n', tensor_pt_dot)
```

```python
tensor_tf_1 = torch.tensor(a)
tensor_tf_2 = torch.tensor(b)

tensor_tf_dot = tf.tensordot(tensor_pt_1, tensor_pt_2, axes=1)

print('Tensor 1:\n', tensor_tf_1)
print('\nTensor 2:\n', tensor_tf_2)
print('\nTích vô hướng 2 tensor:\n', tensor_tf_dot)
```

```python
np.random.seed(2024)
arr_1 = np.random.randint(-10, 10, (3, 3))
arr_2 = np.random.randint(-10, 10, (3, 3))
arr_matmul = np.matmul(arr_1, arr_2)

print('Mảng ngẫu nhiên 1:\n', arr_1)
print('\nMảng ngẫu nhiên 2:\n', arr_2)
print('\nTích 2 matrix:\n', arr_matmul)
```

```python
torch.manual_seed(2024)
tensor_1 = torch.randint(-10, 10, size=(3, 3))
tensor_2 = torch.randint(-10, 10, size=(3, 3))
tensor_matmul = torch.matmul(tensor_1, tensor_2)

print('Tensor ngẫu nhiên 1:\n', tensor_1)
print('\nTensor ngẫu nhiên 2:\n', tensor_2)
print('\nTích 2 matrix:\n', tensor_matmul)
```

```python
tf.random.set_seed(2024)
tensor_1 = tf.random.uniform((3,3), minval=-10, maxval=10, dtype=tf.int32)
tensor_2 = tf.random.uniform((3,3), minval=-10, maxval=10, dtype=tf.int32)
tensor_matmul = tf.matmul(tensor_1, tensor_2)

print('Tensor ngẫu nhiên 1:\n', tensor_1)
print('Tensor ngẫu nhiên 2:\n', tensor_2)
print('\nTích 2 matrix:\n', tensor_matmul)
```
